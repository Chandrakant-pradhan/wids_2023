## Tic-Tac-Toe Simulation
In this assignment you have to construct a python class "TicTacToe" which simulates the game of Tic-Tac-Toe, this will further be used as an enviroment to solve for you in next week.

You may add arguments to given functions as per your need and also additional helper functions (whose name should start with "_", convention for private functions), but only the initially given functions can be called by any user. No additional imports are allowed. The game always starts with X's turn.

```python
class TicTacToe:
    def __init__(self, grid_size):
        board = [['na' for i in range(grid_size)] for j in range(grid_size)]
        self.board = board
        self.turn = 'x'
        self.avail = []
        self.is_over = False
    def act(self, i, j):
        if [i, j] in self.avail or i >= len(self.board) or j >= len(self.board):
            print("NOT allowed")
            return 0
        
        if self.turn == 'x':
            self.board[i][j] = 'x'
            self.turn = 'o'
            self.avail.append([i, j])
        else:
            self.board[i][j] = 'o'
            self.turn = 'x'
            self.avail.append([i, j])
            
        self._IsGameOver()
        return 1
    
    def _IsGameOver(self):
        for row in self.board:
            if all(elem == 'x' for elem in row) or all(elem == 'o' for elem in row):
                self.is_over = True
                return

        for col in range(len(self.board)):
            if all(self.board[row][col] == 'x' for row in range(len(self.board))) or all(self.board[row][col] == 'o' for row in range(len(self.board))):              
                self.is_over = True
                return

        if all(self.board[i][i] == 'x' for i in range(len(self.board))) or all(self.board[i][i] == 'o' for i in range(len(self.board))):              
            self.is_over = True
            return

        if all(self.board[i][len(self.board) - 1 - i] == 'x' for i in range(len(self.board))) or all(self.board[i][len(self.board) - 1 - i] == 'o' for i in range(len(self.board))):              
            self.is_over = True
            return
    
    def reset(self):
        self.board = [['na' for i in range(len(self.board))] for j in range(len(self.board))]
        self.turn = 'x'
        self.avail = []
        self.is_over = False
```

```python
grid_size = int(input("Enter board size : "))
game = TicTacToe(grid_size)

while True:
    current_turn = game.turn
    print(f'{current_turn} turn to play')
    choice = input("want to quit? press q , press anything else to continue :")
    if(choice == 'q'):
        break
    i = int(input("Enter row : "))
    j = int(input("Enter col : "))
     
    
    if not game.act(i, j):
        continue

    for k in range(len(game.board)):
        for l in range(len(game.board[k])):
            print(f'{game.board[k][l]}  ', end='')
        print("\n")
    
    if game.is_over:
        print(f'Game over! winner is {current_turn}')
        break

game.reset()
```
